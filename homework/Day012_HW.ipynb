### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://www.ettoday.net/news/news-list.htm'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text)
# print(soup)
part_list = soup.find('div',{'class':'part_list_2'})
titles = part_list.find_all('a')
for title in titles:
    print(title.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# print(part_list)
dateList = []
typeList = []
nameList = []
h3s = part_list.find_all('h3')
for h3 in h3s:
    dateList.append(h3.find('span').text)
    typeList.append(h3.find('em').text)
    nameList.append(h3.find('a').text)

dic = dict(類別= typeList,時間= dateList,標題= nameList)
pd.set_option('display.max_rows', None)
items = pd.DataFrame(dic)
print(items)
print('===============================')
print('類別:社會')
print('===============================')
boolean = items['類別'].str.contains('社會')
print(items[boolean])
```

### ③ 哪一個類別的文章最多

```python
typeDic = {}
for type in typeList:
    if type not in typeDic.keys():
        typeDic[type] = 0
    else:
        count = typeDic[type] + 1
        typeDic[type] = count
default = ''
for item in typeDic.keys():
    if default != '':
        if typeDic[item] > typeDic[default]:
            default = item
    else:
        default = item
print(default,':',typeDic[default])
```
